크롤링된 pickle 파일 로드

```python
import pickle
data = pickle.load(open("../data/product_info.pkl", "rb"))
```

가격, 카테고리 전처리

```python
import re
for i in range(len(data)):
    if not data[i]["price"].endswith("원"):
        del data[i]["price"]
    else:
        data[i]["price"] = re.sub('[^\d]', "", data[i]["price"])
    
    if len(data[i]["category"]) == 2:
        data[i]["category1"] = data[i]["category"][0]
        data[i]["category2"] = data[i]["category"][1]
    del data[i]["category"]
    del data[i]["url"]
```

데이터프레임으로 변환

```python
import pandas as pd

df = pd.DataFrame(data)
df
```

```python
# id -> product_id로 변환
df.columns = ["product_id", "product_name", "price", "category1", "category2"]
```

```python
df
```

이벤트 항목 제거

```python
df.dropna(subset=["price"], inplace=True)
```

csv 파일 저장

```python
df.to_csv("../data/product_info.csv", index=False)
```
