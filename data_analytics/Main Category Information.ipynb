```python
query = """
SELECT category1 AS 대분류,
       AVG(price) AS 평균가격,
       (SUM(COALESCE(
        CASE
            WHEN sales_1y LIKE '%만 개 이상' THEN CAST(REPLACE(SUBSTR(sales_1y, 1, INSTR(sales_1y, '만') - 1), ',', '') AS DECIMAL(10, 2)) * 10000
            WHEN sales_1y LIKE '%천 개 이상' THEN CAST(REPLACE(SUBSTR(sales_1y, 1, INSTR(sales_1y, '천') - 1), ',', '') AS DECIMAL(10, 2)) * 1000
            WHEN sales_1y LIKE '%개 이상' THEN CAST(REPLACE(SUBSTR(sales_1y, 1, INSTR(sales_1y, '개') - 1), ',', '') AS DECIMAL(10, 2))
            ELSE NULL
        END, 0) * price) || '') AS 연간총매출,
        AVG(COALESCE(
        CASE
            WHEN view_1m LIKE '%만 회 이상' THEN CAST(SUBSTR(view_1m, 1, INSTR(view_1m, '만') - 1) AS DECIMAL(10, 2)) * 10000
            WHEN view_1m LIKE '%천 회 이상' THEN CAST(SUBSTR(view_1m, 1, INSTR(view_1m, '천') - 1) AS DECIMAL(10, 2)) * 1000
            WHEN view_1m LIKE '%회 이상' THEN CAST(SUBSTR(view_1m, 1, INSTR(view_1m, '회') - 1) AS DECIMAL(10, 2))
            ELSE NULL
        END, 0)) AS 월간평균조회수,
       AVG(sales_1y) AS 연간평균판매수,
       AVG(COALESCE(
        CASE
            WHEN sales_1y LIKE '%만 개 이상' THEN CAST(SUBSTR(sales_1y, 1, INSTR(sales_1y, '만') - 1) AS DECIMAL(10, 2)) * 10000
            WHEN sales_1y LIKE '%천 개 이상' THEN CAST(SUBSTR(sales_1y, 1, INSTR(sales_1y, '천') - 1) AS DECIMAL(10, 2)) * 1000
            WHEN sales_1y LIKE '%개 이상' THEN CAST(SUBSTR(sales_1y, 1, INSTR(sales_1y, '개') - 1) AS DECIMAL(10, 2))
            ELSE NULL
        END, 0)) / (AVG(COALESCE(
        CASE
            WHEN view_1m LIKE '%만 회 이상' THEN CAST(SUBSTR(view_1m, 1, INSTR(view_1m, '만') - 1) AS DECIMAL(10, 2)) * 10000
            WHEN view_1m LIKE '%천 회 이상' THEN CAST(SUBSTR(view_1m, 1, INSTR(view_1m, '천') - 1) AS DECIMAL(10, 2)) * 1000
            WHEN view_1m LIKE '%회 이상' THEN CAST(SUBSTR(view_1m, 1, INSTR(view_1m, '회') - 1) AS DECIMAL(10, 2))
            ELSE NULL
        END, 0)) * 12) AS 조회수대비판매수,
       AVG(REPLACE(sale_m, '%', '')) AS 남성구매자비율평균,
       AVG(REPLACE(sale_f, '%', '')) AS 여성구매자비율평균
FROM product
WHERE category1 IS NOT NULL
GROUP BY category1;
"""
main = psql.sqldf(query, locals())
main['연간총매출'] = main['연간총매출'].apply(lambda x: "{:,.0f}".format(float(x)))
main
```

```python
for col in main.columns[1:]:
    # 문자열 데이터만 처리
    if main[col].dtype == 'object':
        main[col] = pd.to_numeric(main[col].str.replace(',', ''), errors='coerce')

    # 숫자형 데이터를 소수점 셋째자리까지로 변환
    if main[col].dtype == 'float':
        main[col] = main[col].round(3)


main
```

```python
main.to_csv('Comparative_Information_by_Main_Category.csv', index=False)
```
